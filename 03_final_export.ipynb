```python
import pandas as pd
import numpy as np
import sqlite3
from sqlalchemy import create_engine
import os

BASE         = 'C:/Users/Saurabh yadav/Documents/Patient Readmission Risk Analysis/'
TABLEAU_PATH = BASE + 'tableau/'
DB_PATH      = BASE + 'database/healthcare.db'

os.makedirs(BASE + 'database/', exist_ok=True)

df = pd.read_csv(TABLEAU_PATH + 'clean_patients.csv')
print('Data loaded:', df.shape)
```

```python
df['risk_score'] = (
    (df['num_medications'] / df['num_medications'].max() * 0.3) +
    (df['time_in_hospital'] / df['time_in_hospital'].max() * 0.3) +
    (df['total_visits'] / (df['total_visits'].max() + 1) * 0.4)
).round(4)

def risk_label(score):
    if score >= 0.6:   return 'High Risk'
    elif score >= 0.3: return 'Medium Risk'
    else:              return 'Low Risk'

df['risk_label'] = df['risk_score'].apply(risk_label)

engine = create_engine('sqlite:///' + DB_PATH)
df.to_sql('patients', engine, if_exists='replace', index=False)

print('Risk scores added')
print('Data saved to database')
print()
print('Risk distribution:')
print(df['risk_label'].value_counts())
```

```python
final_df = df.copy()
final_df.to_csv(TABLEAU_PATH + 'final_patients_tableau.csv', index=False)

print('Final Tableau CSV exported!')
print('File:', TABLEAU_PATH + 'final_patients_tableau.csv')
print('Rows:', len(final_df))
print('Columns:', len(final_df.columns))
print()
print('Columns in file:')
for col in final_df.columns:
    print(' -', col)
print()
print('Phase 3 Complete!')
print('Now open Tableau and load: final_patients_tableau.csv')
```
